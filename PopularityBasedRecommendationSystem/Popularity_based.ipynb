```python
import numpy as np
import pandas as pd
```

```python
i_cols = ['movie id', 'movie title' ,'release date','video release date', 'IMDb URL', 'unknown', 'Action', 'Adventure',
           'Animation', 'Children\'s', 'Comedy', 'Crime', 'Documentary', 'Drama', 
          'Fantasy', 'Film-Noir', 'Horror', 'Musical', 'Mystery', 'Romance', 'Sci-Fi', 'Thriller', 'War', 'Western']
items = pd.read_csv('ml-100k/u.item', sep='|', names=i_cols,encoding='latin-1')
```

```python
#Select only needed Item
movie_content=items[['movie id','movie title','Action', 'Adventure',
                     'Animation', 'Children\'s', 'Comedy', 'Crime', 'Documentary', 'Drama', 'Fantasy','Film-Noir', 'Horror', 
                     'Musical', 'Mystery', 'Romance', 'Sci-Fi', 'Thriller', 'War', 'Western']]
```

```python
movie_content
```

```python
#read the ratings dataset
r_cols = ['user_id', 'movie_id', 'rating', 'unix_timestamp']
ratings = pd.read_csv('ml-100k/u.data', sep='\t', names=r_cols,encoding='latin-1')
```

```python
ratings
```

```python
#Calculate the number of ratings for each movie
popularity = ratings.groupby('movie_id')['rating'].count().reset_index()
```

```python
popularity
```

```python
popularity.columns = ['movie id','rating']
popularity
```

```python
# Sort movies by rating for each genre
genre = ['Action', 'Adventure', 'Animation', 'Children\'s', 'Comedy', 'Crime', 'Documentary', 'Drama',
                 'Fantasy', 'Film-Noir', 'Horror', 'Musical', 'Mystery', 'Romance', 'Sci-Fi', 'Thriller', 'War', 'Western']
```

```python
genre
```

```python
#check for action genre

genre_movies = movie_content[movie_content['Action'] == 1]
```

```python
genre_movies
```

```python
#merge the dataframes for popularity information
pop_movies = pd.merge(genre_movies,popularity, on='movie id',how='left')
```

```python
pop_movies
```

```python
Recommended_movies =  pop_movies.sort_values(by='rating',ascending=False).head(5)
```

```python
Recommended_movies
```

```python
Recommended_movies=Recommended_movies[['movie title','rating']]
```

```python
Recommended_movies
```

```python
#library created
genre_recommendations = {}
for gen in genre:
    #Filter movies in the respective genre
    genre_movies = movie_content[movie_content[gen] == 1]
    
    #Merge with popularity data
    genre_movies = pd.merge(genre_movies, popularity, on='movie id', how='left')
    
    #Sort by rating(popularity) and select the top 5
    Recommended_movies =  genre_movies.sort_values(by='rating',ascending=False).head(5)
    
    #Store the movie title and rating for the respective genre in genre_recommendations library
    #storing the movies genre and its corresponding movie title and rating
    genre_recommendations[gen] = Recommended_movies[['movie title', 'rating']] 

    #Create a DataFrame to store the recommended movies for each genre
    recommended_df = pd.DataFrame()
    
#passing genre and getting its corresponding top 5 movie title
for gen, recommendations in genre_recommendations.items():
    recommended_df[gen] = recommendations['movie title'].values 
popular_movies = recommended_df
```
